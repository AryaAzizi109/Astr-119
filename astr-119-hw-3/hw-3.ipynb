```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 1000)


n = 5.5*np.cos(2*x) + 5.5
w = 0.02*np.exp(x)
a = 0.25*np.square(x) + 0.1*np.sin(10*x)


#plot n w a 
plt.plot(x,n)
plt.plot(x,w)
plt.plot(x,a)


#change axis
plt.xlim(0,2*np.pi)
plt.ylim(-1,10)


#change axis name
plt.ylabel("measures of Awesomeness")
plt.xlabel("Time in Astr 119")

#show plot
plt.show()
```
